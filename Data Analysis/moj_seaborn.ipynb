```python
import seaborn as sns
import pandas as pd
%matplotlib inline
```

```python
data = pd.read_csv("dane_csv.csv")
```

```python
data
```

```python
sns.distplot(data['wzrost'], kde = True)
```

```python
sns.boxplot(data = data, x = 'wzrost')
```

```python
sns.violinplot(data = data, x = 'wzrost')
```

```python
sns.violinplot(data = data, y = 'wzrost', x = 'płeć', split = True, palette={"K": "purple", "M": "royalblue"})
```

```python
sns.violinplot(data = data, x = 'kolor oczu', y = 'wzrost', hue = 'płeć', split = False, palette={"K": "purple", "M": "royalblue"})
```

```python
sns.violinplot(data = data, x = 'kolor oczu', y = 'wzrost', hue = 'płeć', split = True, palette={"K": "purple", "M": "royalblue"})
```

```python
sns.violinplot(data = data, x = 'wzrost', y = 'płeć', palette = 'hot')
sns.swarmplot(data = data, x = 'wzrost', y = 'płeć', color = 'black')
```

```python
sns.violinplot(data = data, x = 'wzrost')
sns.swarmplot(data = data, x = 'wzrost', color = 'black')
```

```python
sns.violinplot(data = data, x = 'wzrost', inner = "quartile")
```

```python
sns.violinplot(data = data, x = 'wzrost', inner = "stick")
```

```python
sns.violinplot(data = data, x = 'wzrost', inner = "point")
```

```python
sns.violinplot(data = data, x = 'wzrost', y = 'płeć', inner = 'quartile')
```
